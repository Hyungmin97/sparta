```python
import random

#1부터 10 사이의 랜덤 숫자 생성
random_number = random.randint(1,10)
#랜덤 숫자 출력 확인
#print(random_number)

#반복문 생성
while True:
    #플레이어 숫자 입력
    player_input = int(input("1~10 사이의 숫자를 입력하세요: "))
    #조건문 생성
    if random_number == player_input:
        print(f"정답입니다. 랜덤 숫자: {random_number}, 입력 숫자: {player_input}")
        break
    elif random_number > player_input:
        print(f"숫자가 너무 작습니다. 입력 숫자: {player_input}")
    else:
        print(f"숫자가 너무 큽니다. 입력 숫자: {player_input}")
```

```python
#2. 클래스와 함수 사용하기
# 이름, 성별, 나이를 입력받고, 이를 출력하는 프로그램을 작성하시오.
```

```python
#클래스 정의
class Person:
    def __init__(self, name, gender, age):
        self.name = name
        self.gender = gender
        self.age = age
    def display(self):
        print(f"이름: {self.name}, 성별: {self.gender}")
        print(f"나이: {self.age}")
    
#정보 입력
name = str(input("이름을 입력하세요: "))
gender = str(input("성별을 입력하세요(male/female): "))
age = int(input("나이를 입력하세요: "))

#입력 받은 값으로 Person 객체 생성
person = Person(name, gender, age)

#객체의 정보 출력
person.display()
```

```python
#3. python 라이브러리로 데이터 분석하기
#python 라이브러리를 활용하여 주어진 데이터(.xlsx)를 분석 Quiz를 수행해주세요.

#과제  수행을 위한 데이터
#"관서별 5대범죄 발생 및 검거.xlsx"
#"pop_kor.csv"

#1. Python 라이브러리 함수를 사용하여 엑셀 파일을 불러오고, DataFrame을 출력해주세요.
#2. 각 경찰서(관서명)를 해당 구 이름으로 매핑하여 '구별'이라는 새로운 column을 생성하고, DataFrame을 출력해주세요.
#    - 매칭되지 않는 경찰서명에 대해서는 기본값 '구 없음'을 할당합니다.
#3. pivot_table 을 사용하여 관서별 데이터를 구별 데이터로 변경하고, 같은 구의 경우에는 sum을 적용하여 더해주세요. (index : 관서 이름 -> 구 이름)
#4. 구 없음  행은 drop 을 활용하여 삭제해주세요.
#5. 각 범죄 별로 검거율을 계산하고, 각 검거율 데이터 column을 DataFrame에 추가해주세요.
#6. 필요없는 column을 del 을 사용하여 삭제해주세요.
#7. DataFrame의 컬럼명을 rename 을 사용하여 변경해주세요.
```

```python
import pandas as pd

# 엑셀 파일과 CSV 파일 불러오기
excel_data = pd.read_excel('관서별 5대범죄 발생 및 검거.xlsx')
csv_data = pd.read_csv('pop_kor.csv')
```

```python
# DataFrame 출력
print(excel_data)
```

```python
#경찰서명을 구 이름으로 매핑하는 dictionary 생성
gu_data = {
    '중부서': '중구',
    '종로서': '종로구',
    '남대문서': '용산구',
    '서대문서': '서대문구',
    '혜화서': '종로구',
    '용산서': '용산구',
    '성북서': '성북구',
    '동대문서': '동대문구',
    '마포서': '마포구',
    '영등포서': '영등포구',
    '성동서': '성동구',
    '동작서': '동작구',
    '광진서': '광진구',
    '서부서': '은평구',
    '강북서': '강북구',
    '금천서': '금천구',
    '중랑서': '중랑구',
    '강남서': '강남구',
    '관악서': '관악구',
    '강서서': '강서구',
    '강동서': '강동구',
    '종암서': '성북구',
    '구로서': '구로구',
    '서초서': '서초구',
    '양천서': '양천구',
    '송파서': '송파구',
    '노원서': '노원구',
    '방배서': '서초구',
    '은평서': '은평구',
    '도봉서': '도봉구',
    '수서서': '강남구'
}
```

```python
#'구별' 칼람 생성, 매칭되지 않는 경찰서는 기본값 '구 없음' 할당.
excel_data['구별'] = excel_data['관서명'].map(gu_data).fillna('구 없음')
```

```python
#DataFrame 출력
print(excel_data)
```

```python
#pivot_table을 사용하여 구별로 데이터 합산 (sum 적용)
crime_by_district = pd.pivot_table(excel_data, index = '구별', aggfunc='sum')
```

```python
# DataFrame 출력
print(crime_by_district)
```

```python
# '구 없음' 행 삭제
crime_by_district = crime_by_district.drop('구 없음', errors='ignore')
```

```python
# 범죄 별 검거율 계산 (검거율 = 검거 / 발생 * 100)
crime_by_district['살인 검거율'] = crime_by_district['살인(검거)'] / crime_by_district['살인(발생)'] * 100
crime_by_district['강도 검거율'] = crime_by_district['강도(검거)'] / crime_by_district['강도(발생)'] * 100
crime_by_district['강간 검거율'] = crime_by_district['강간(검거)'] / crime_by_district['강간(발생)'] * 100
crime_by_district['절도 검거율'] = crime_by_district['절도(검거)'] / crime_by_district['절도(발생)'] * 100
crime_by_district['폭력 검거율'] = crime_by_district['폭력(검거)'] / crime_by_district['폭력(발생)'] * 100
```

```python
# DataFrame 출력
print(crime_by_district)
```

```python
#필요 없는 column 삭제
columns_to_delete = ['살인(검거)', '강도(검거)', '강간(검거)', '절도(검거)', '폭력(검거)']
for col in columns_to_delete:
    del crime_by_district[col]
```

```python
# DataFrame 출력
print(crime_by_district)
```

```python
# 컬럼명 변경
crime_by_district.rename(columns={
    '살인(발생)': '살인 사건수',
    '강도(발생)': '강도 사건수',
    '강간(발생)': '강간 사건수',
    '절도(발생)': '절도 사건수',
    '폭력(발생)': '폭력 사건수'
}, inplace=True)
```

```python
# 최종 DataFrame 출력
print(crime_by_district)
```
